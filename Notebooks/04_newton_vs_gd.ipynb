### Summary :

# Newton vs GD - Results Summary

## Convergence Speed

| Dataset | GD Iterations | Newton Iterations |
|---------|--------------|-------------------|
| Breast Cancer | 1000 | 11 |
| Diabetes | 1000 | 6 |
| Heart Disease | 1000 | 7 |

## Key Finding

Both optimizers reach the same final performance on all 3 datasets.
Newton gets there in a fraction of the iterations.

This is because logistic regression is a convex problem — there is only
one global minimum. Both methods find it, just at different speeds.

Newton trades higher computation per iteration (Hessian inversion)
for dramatically fewer iterations overall. On small datasets this is
a clear win. On large datasets like Credit Fraud or Adult Income,
the cost of inverting a massive Hessian makes Newton impractical.

## Visual Evidence

GD loss curve is a smooth gradual descent over 1000 steps.
Newton drops almost vertically to the minimum in ~11 steps.
Same endpoint, completely different journey.
