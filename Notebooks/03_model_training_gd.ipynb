## What the Results Tell Us

Breast Cancer and Heart Disease are the strongest performers with MCC above 0.71.
These datasets are well balanced and clean — GD handles them well.

Credit Fraud and Stroke look impressive on accuracy but MCC exposes the truth.
Both are near zero meaning the model learned nothing useful — it just predicts
the majority class every time. Class imbalance completely defeats plain GD here.

Diabetes and Adult Income sit in honest middle ground. Real predictive power
but room for improvement with regularization and threshold tuning.

## Key Lesson

Never trust accuracy alone on imbalanced datasets.
MCC is the only metric that tells the full story.
