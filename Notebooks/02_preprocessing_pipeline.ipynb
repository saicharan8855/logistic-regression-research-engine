# Summary :

# What We Did in Preprocessing

The goal of this notebook was to take all six raw datasets and transform them 
into clean, model-ready arrays ready for training. Here is exactly what happened 
to each dataset.

---

## The Pipeline (Same Order for Every Dataset)

1. Loaded raw CSV files from `Data/raw/`
2. Dropped useless columns (id, empty columns)
3. Fixed hidden zeros in Diabetes (replaced with NaN)
4. Separated features X and target y
5. Shuffled data with seed=11
6. Binary encoded all targets to {0, 1}
7. One-hot encoded categorical columns (Stroke and Adult Income only)
8. Split into 80% train / 20% test
9. Imputed missing values using median (fit on train only)
10. Removed multicollinear features with correlation > 0.9 (Breast Cancer only)
11. Standardized features (fit scaler on train, applied to both)
12. Added bias column (ones at index 0)
13. Checked for perfect separation (diagnostic only)
14. Saved all arrays to `Data/processed/`

---

## What Changed Per Dataset

**Breast Cancer**
- Dropped `id` and `Unnamed: 32` (fully empty column)
- VIF check revealed severe multicollinearity in almost all features
- Features reduced from 30 → 18 after removing correlated pairs above 0.9
- Final shape: train (455, 19) | test (114, 19)

**Credit Fraud**
- No categorical columns, no missing values, no multicollinearity
- Standard pipeline only
- Final shape: train (227845, 31) | test (56962, 31)

**Heart Disease**
- Cleanest dataset — no issues found
- Standard pipeline only
- Final shape: train (820, 14) | test (205, 14)

**Diabetes**
- Replaced impossible zeros in Glucose, BloodPressure, SkinThickness, 
  Insulin, BMI with NaN before imputation
- Insulin had 374 hidden nulls — highest among all datasets
- Final shape: train (614, 9) | test (154, 9)

**Stroke**
- Dropped `id` column
- Imputed 201 missing BMI values with median
- One-hot encoded 5 categorical columns → features expanded from 10 → 21
- Final shape: train (4088, 22) | test (1022, 22)

**Adult Income**
- One-hot encoded 8 categorical columns → features expanded from 14 → 108
- native.country alone contributed most of the expansion
- Final shape: train (26048, 109) | test (6513, 109)

---

## Key Decisions Made

- **Median imputation** over mean — more robust to outliers
- **Fit scaler on train only** — prevents data leakage from test set
- **Multicollinearity removal on train only** — same keep indices applied to test
- **No perfect separation detected** in any dataset — safe to train normally
- **seed=11** used everywhere for full reproducibility

---

## Output

24 numpy arrays saved to `Data/processed/` —
X_train, X_test, y_train, y_test for all 6 datasets.
These files are loaded directly in all future training notebooks.
No preprocessing needs to be repeated.
